```python
import numpy as np
import matplotlib.pyplot as plt
pi=np.pi
x=np.linspace(-4*pi, 4*pi, 1000)

plt.plot(x, (np.sin(x)/x)**2)
plt.show()
plt.plot(x, np.sin(x)/x**2)
plt.show()

x=np.linspace(-1, 40, 1000)
plt.plot(x, 5*x+6)
plt.plot(x, x**2+3*x+6)
plt.show()


x=np.linspace(0, 10, 10)
plt.plot(x, x**0+9)
x=np.linspace(10, 20, 10)
plt.plot(x, x**0+19)
x=np.linspace(20, 30, 10)
plt.plot(x, x**0+29)

coord = [[20,10], [25,25], [17,20], [10,10], [20,15]]
coord.append(coord[0]) #repeat the first point to create a 'closed loop'

xs, ys = zip(*coord) #create lists of x and y values

plt.figure()
plt.plot(xs,ys)
plt.show()

x=np.linspace(-11, 9, 100)
plt.plot(x, x**2+8*x+12)
plt.plot(x, 4*x+8)
plt.show()
```

```python
plt.style.use('fivethirtyeight')
x=np.linspace(-1, 8, 10000)
plt.plot(x, 6*x + x**2, linewidth=1)
plt.plot(x, x**2-4*x+20, linewidth=1)
plt.plot(x, 6*x + x**2 + x**2-4*x+20)
plt.show()
```
