# Create A Candle Stick Chart

```python
import pandas as pd
from pandas_datareader import data as web
import plotly.graph_objects as go
```

```python
stock = 'ABT'
 
df = web.DataReader(stock, data_source='yahoo', start='01-01-2020')
df
```

```python
trace1 = {
    'x': df.index,
    'open': df.Open,
    'close': df.Close,
    'high': df.High,
    'low': df.Low,
    'type': 'candlestick',
    'name': 'Abbot',
    'yaxis': 'y2',
    # increasing_line_color='green',
    # decreasing_line_color='red',
    'showlegend': True
}
```

```python
# Calculate and define moving average of 20 periods
avg_20 = df.Close.rolling(window=20, min_periods=1).mean()

# Calculate and define moving average of 60 periods
avg_60 = df.Close.rolling(window=60, min_periods=1).mean()
```

```python
trace2 = {
    'x': df.index,
    'y': avg_20,
    'yaxis': 'y2',
    'type': 'scatter',
    'mode': 'lines',
    'line': {
        'width': 1,
        'color': 'purple'
            },
    'name': 'Moving Average of 20 periods'
}
```

```python
trace3 = {
    'x': df.index,
    'y': avg_60,
    'yaxis': 'y2',
    'type': 'scatter',
    'mode': 'lines',
    'line': {
        'width': 1,
        'color': 'red'
    },
    'name': 'Moving Average of 60 periods'
}
```

```python
colors = []
INCREASING_COLOR = '#17BECF'
DECREASING_COLOR = '#7F7F7F'

for i in range(len(df.Close)):
    if i != 0:
        if df.Close[i] > df.Close[i-1]:
            colors.append(INCREASING_COLOR)
        else:
            colors.append(DECREASING_COLOR)
    else:
        colors.append(DECREASING_COLOR)
```

```python
trace4 = {
    'x': df.index,
    'y': df.Volume,
    'marker': dict( color=colors ),
    'type': 'bar', 
    'yaxis': 'y', 
    'name': 'Volume'
}
```

```python
data = [trace1, trace2, trace3, trace4]
# Config graph layout
layout = go.Layout({
    'title': {
        'text': 'ABBOTT OHLC and SMA20, 60',
        'font': {
            'size': 12
        }
    },
    "yaxis": {"domain": [0, 0.2]},
    "yaxis2": {"domain": [0.2, 0.8]}
})
```

```python
fig = go.Figure(data=data, layout=layout)
fig.show()
```
