```python
from fastquant import backtest, get_stock_data
abt = get_stock_data('ABT', "2018-01-01", "2019-01-01" )
print(abt.head())
backtest('smac', abt, fast_period=15, slow_period=40)
```

```python
tsla = get_stock_data("TSLA", "2019-01-01", "2020-09-10", format="ohlcv")
print(tsla.head())
```

```python
backtest("smac", tsla, buy_prop=0.50, sell_prop=0.50, commission=0.01)
```
